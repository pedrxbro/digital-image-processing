```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage.util import random_noise
```

```python
def load_image(img_name, color):
    # Mapeamento dos nomes para caminhos
    paths = {
        'a': 'data/Fig0333(a)(test_pattern_blurring_orig).tif',
        'moon': 'data/Fig0338(a)(blurry_moon).tif',
        'dip-xe': 'data/Fig0340(a)(dipxe_text).tif',
        'circle': 'data/Fig0342(a)(contact_lens_original).tif'
    }

    # Define flag do OpenCV
    if color == 1:
        flag = cv2.IMREAD_COLOR  # BGR
    else:
        flag = cv2.IMREAD_GRAYSCALE  # escala de cinza

    return cv2.imread(paths[img_name], flag)
```

```python
def show_image(img_name, color):
    img = load_image(img_name, color)

    cv2.imshow('in',img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
#Conversão básica

def grayscale_basic(img):
    B, G, R = cv2.split(img)
    img_grayscale_basic = (B + G + R)/3

    img_grayscale_basic = np.array(img_grayscale_basic, dtype = np.uint8)

    cv2.imshow('img_grayscale_basic',img_grayscale_basic)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
#Conversão ponderada

def grayscale_pondered(img):
    B, G, R = cv2.split(img)
    img_grayscale_pondered = 0.299*R + 0.587*G + 0.114*B

    img_grayscale_pondered = np.array(img_grayscale_pondered, dtype=np.uint8)

    return img_grayscale_pondered
```

```python
#Função Gauss Kernel

def gauss_create(sigma=1, size_x=3, size_y=3):
    x, y = np.meshgrid(np.linspace(-1,1,size_x), np.linspace(-1,1,size_y))
    calc = 1/(2*np.pi*(sigma**2))
    exp = np.exp(-(((x**2) + (y**2))/(2*(sigma**2))))
    kernel = exp * calc

    return kernel / np.sum(kernel)
```

```python
#Função kernel média

def mean_kernel(size):
    kernel = np.ones((size, size), dtype=float)
    kernel = kernel / np.sum(kernel)
    return kernel
```

```python
# Função adiciona borda

def add_padding(img, padding_height, padding_width):
    n, m = img.shape

    padded_img = np.zeros((n + padding_height * 2, m + padding_width * 2))
    padded_img[padding_height : n + padding_height, padding_width : m + padding_width] = img

    return padded_img
```

```python
#Função Kernel Mediana

def median_filter(img, size, padding=True):
    img_height, img_width = img.shape
    pad = size // 2

    if padding == True:
        padded_img = add_padding(img, pad, pad)
    else:
        padded_img = img

    output = np.zeros((img_height, img_width), dtype=float)

    for i in range(img_height):
        for j in range(img_width):
            region = padded_img[i:i + size, j:j + size]
            values = region.flatten()
            values.sort()
            median_value = values[len(values) // 2]
            output[i, j] = median_value

    return np.array(output, dtype=np.uint8)
```

```python
#Função de convolução

def conv2d(img, kernel, padding=True):
    # Get dimensions of the kernel
    k_height, k_width = kernel.shape  # Atribui valor à variável k_height, k_width

    # Get dimensions of the image
    img_height, img_width = img.shape  # Atribui valor à variável img_height, img_width

    # Calculate padding required
    pad_height = k_height // 2  # Atribui valor à variável pad_height
    pad_width = k_width // 2  # Atribui valor à variável pad_width

    # Create a padded version of the image to handle edges
    if padding == True:
        padded_img = add_padding(img, pad_height, pad_width)  # Atribui valor à variável padded_img

    # print(padded_img)

    # Initialize an output image with zeros
    output = np.zeros((img_height, img_width), dtype=float)  # Atribui valor à variável output


    # Perform convolution (Implementação)
    for i in range(img_height):
        for j in range(img_width):
            region = padded_img[i:i + k_height, j:j + k_width]
            value = np.sum(region * kernel)
            output[i, j] = value

    return output
```

## Observações sobre os filtros de suavização

De acordo com os dois testes realizados (sem adição de ruído e com adição), podemos observar a seguinte situação:

### Filtro da Média
- **Comportamento visual:** Suaviza a imagem uniformemente, mas tende a **borrar bordas e detalhes finos**, especialmente com kernels maiores (5x5 e 7x7).
- **Ruído:** Remove parcialmente ruídos leves, mas não é eficaz contra ruídos impulsivos ("sal e pimenta").
- **Esmaecimento:** Quanto maior o kernel, mais a imagem parece "esmaecida" ou desfocada, pois cada pixel é substituído pela média dos vizinhos.

### Filtro Gaussiano
- **Comportamento visual:** Produz uma suavização mais natural que a média, **preservando melhor as bordas**, pois dá mais peso aos pixels próximos do centro do kernel.
- **Ruído:** Reduz ruídos leves de forma eficaz, mas ainda pode suavizar detalhes pequenos.
- **Esmaecimento:** O aumento do sigma ou do tamanho do kernel aumenta o efeito de desfoque, mas de maneira mais gradual e perceptualmente agradável que a média.

### Filtro da Mediana
- **Comportamento visual:** Mantém bordas e estruturas mais nítidas mesmo com kernels maiores.
- **Ruído:** Excelente para remover **ruídos impulsivos**, sem borrar demais a imagem.
- **Esmaecimento:** Comparado à média e ao gaussiano, a mediana causa menos esmaecimento; a imagem permanece visualmente mais definida, mesmo para kernels 5x5 e 7x7.

### Resumo do comportamento
- **Média:** suavização uniforme, borramento significativo das bordas, boa para ruído leve.
- **Gaussiano:** suavização ponderada, preserva bordas melhor que a média, reduz ruído leve com aparência natural.
- **Mediana:** preserva bordas, excelente contra ruído impulsivo, mantém a definição da imagem.

Em geral, o tamanho do kernel amplifica os efeitos descritos: **kernels maiores aumentam a suavização e o esmaecimento**, enquanto kernels menores mantém mais detalhes.

```python
#Função para normalizar mínimo e máximo
def normalize_minmax(img):
    img = img.astype(np.float32)
    min_val = img.min()
    max_val = img.max()

    if max_val == min_val:
        return np.zeros_like(img, dtype=np.uint8)

    norm = (img - min_val) * 255.0 / (max_val - min_val)
    return norm.astype(np.uint8)
```

```python
#Kernel
laplacian_1 = np.array([
    [0, 1, 0],
    [1, -4, 1],
    [0, 1, 0]
], dtype=float)

laplacian_2 = np.array([
    [1, 1, 1],
    [1, -8, 1],
    [1, 1, 1]
], dtype=float)

laplacian_3 = np.array([
    [0, -1, 0],
    [-1, 4, -1],
    [0, -1, 0]
], dtype=float)

laplacian_4 = np.array([
    [-1, -1, -1],
    [-1, 8, -1],
    [-1, -1, -1]
], dtype=float)
```

```python
#Função Laplacian
def laplacian_sharpen(img, kernel):
    laplacian = conv2d(img, kernel)

    if kernel[kernel.shape[0]//2, kernel.shape[1]//2] < 0:
        sharpen = img.astype(np.float32) - laplacian
    else:
        sharpen = img.astype(np.float32) + laplacian

    sharpen = np.clip(sharpen, 0, 255)
    return laplacian, sharpen.astype(np.uint8)
```

```python
#Monta as imagens Laplacian
img = load_image('moon', 0)

kernels = {
    "Laplaciano 4-vizinhos": laplacian_1,
    "Laplaciano 8-vizinhos": laplacian_2,
    "Laplaciano 4-vizinhos invertido": laplacian_3,
    "Laplaciano 8-vizinhos invertido": laplacian_4
}

plt.figure(figsize=(20, 10))

# Coluna 1 = imagem original
for row in range(3):
    plt.subplot(3, 5, row*5 + 1)
    plt.imshow(img, cmap='gray')
    plt.title("Original")
    plt.axis('off')

# Colunas 2..5 = kernels
for i, (name, kernel) in enumerate(kernels.items()):
    laplacian, sharpen = laplacian_sharpen(img, kernel)

    # linha 1 -> Laplaciano sem scaling
    lap_display = np.clip(laplacian, 0, 255).astype(np.uint8)
    plt.subplot(3, 5, i + 2)
    plt.imshow(lap_display, cmap='gray')
    plt.title(f"{name}\nSem scaling")
    plt.axis('off')

    # linha 2 -> Laplaciano com scaling
    plt.subplot(3, 5, i + 7)
    plt.imshow(normalize_minmax(laplacian), cmap='gray')
    plt.title(f"{name}\nCom scaling")
    plt.axis('off')

    # linha 3 -> Sharpen
    plt.subplot(3, 5, i + 12)
    plt.imshow(sharpen, cmap='gray')
    plt.title(f"{name}\nSharpen")
    plt.axis('off')

plt.tight_layout()
plt.show()
```

## Observações — Laplaciano (com scaling, sem scaling e sharpening)

A aplicação do operador Laplaciano com diferentes formas de visualização (com e sem scaling) e sua combinação com a imagem original (sharpening) evidencia diferenças importantes no comportamento visual:

- **Laplaciano sem scaling:**
  - **Comportamento visual:** A imagem resultante apresenta valores negativos e positivos, o que pode gerar uma visualização muito escura ou com pouco contraste quando exibida diretamente.
  - **Bordas:** As bordas estão presentes, mas muitas vezes difíceis de interpretar visualmente devido à faixa de valores não normalizada.
  - **Ruído:** O ruído aparece de forma intensa, porém pode ficar “escondido” pela má visualização.
  - **Observação:** Sem normalização, o resultado não é ideal para análise visual direta.

- **Laplaciano com scaling (normalização):**
  - **Comportamento visual:** Após aplicar o scaling (normalização para faixa visível, como 0–255), as bordas tornam-se claramente visíveis.
  - **Bordas:** Evidenciadas com alto contraste, destacando contornos e detalhes finos.
  - **Ruído:** Fica muito mais perceptível, pois o scaling também realça pequenas variações de intensidade.
  - **Observação:** O scaling não altera a informação, apenas melhora a visualização — tornando explícito tanto bordas quanto ruídos.

- **Sharpening (imagem original − Laplaciano):**
  - **Comportamento visual:** A imagem resultante apresenta aumento significativo de nitidez, com contornos mais definidos.
  - **Bordas:** São reforçadas diretamente na imagem original, melhorando a percepção de detalhes.
  - **Ruído:** Também é amplificado, principalmente em regiões homogêneas.
  - **Esmaecimento:** Reduz o efeito de desfoque, deixando a imagem mais “crisp”.
  - **Observação:** O sharpening com Laplaciano é eficaz para realce, mas pode introduzir artefatos se aplicado de forma intensa.

### Resumo do comportamento
- **Sem scaling:** difícil visualização, apesar de conter a informação das bordas.
- **Com scaling:** bordas e ruídos ficam evidentes e analisáveis.
- **Sharpening:** melhora a nitidez geral, mas amplifica ruído junto com os detalhes.

```python
#Função Unsharp Masking
def unsharp_masking(img, blur_kernel):
    img = img.astype(np.float32)
    blurred = conv2d(img, blur_kernel).astype(np.float32)

    mask = img - blurred
    sharpened = img + mask

    sharpened = np.clip(sharpened, 0, 255).astype(np.uint8)
    return blurred, mask, sharpened
```

```python
#Monta as Imagens unsharp_masking
img = load_image('dip-xe', 0)

kernels_blur = {
    "Média 3x3": mean_kernel(3),
    "Média 5x5": mean_kernel(5),
    "Média 7x7": mean_kernel(7)
}

plt.figure(figsize=(16, 10))
#Plot da original
for row in range(3):
    plt.subplot(3, 4, row*4 + 1)
    plt.imshow(img, cmap='gray', vmin=0, vmax=255)
    plt.title("Original")
    plt.axis('off')



for i, (name, kernel) in enumerate(kernels_blur.items()):
    blurred, mask, sharpened = unsharp_masking(img, kernel)

    # linha 1 -> borrada
    plt.subplot(3, 4, i + 2)
    plt.imshow(blurred.astype(np.uint8), cmap='gray')
    plt.title(f"{name}\nBorrada")
    plt.axis('off')

    # linha 2 -> máscara de detalhes
    plt.subplot(3, 4, i + 6)
    plt.imshow(normalize_minmax(mask), cmap='gray')
    plt.title(f"{name}\nMáscara")
    plt.axis('off')

    # linha 3 -> resultado final
    plt.subplot(3, 4, i + 10)
    plt.imshow(sharpened, cmap='gray', vmin=0, vmax=255)
    plt.title(f"{name}\nUnsharp")
    plt.axis('off')

plt.tight_layout()
plt.show()
```

### Highboost Filtering

```python
#Função highboost filtering

def highboost_filter(img, blur_kernel, k=1.5):
    img = img.astype(np.float32)
    blurred = conv2d(img, blur_kernel).astype(np.float32)

    mask = img - blurred
    sharpened = img + k * mask

    sharpened = np.clip(sharpened, 0, 255).astype(np.uint8)
    return blurred, mask, sharpened
```

```python
#Monta as imagens
img = load_image('dip-xe', 0)

factors = [1.0, 1.5, 2.0, 3.0]

plt.figure(figsize=(18, 8))

for row in range(2):
    plt.subplot(2, 5, row*5 + 1)
    plt.imshow(img, cmap='gray', vmin=0, vmax=255)
    plt.title("Original")
    plt.axis('off')

for i, k in enumerate(factors):
    _, mask, sharpened = highboost_filter(img, mean_kernel(3), k=k)

    plt.subplot(2, 5, i + 2)
    plt.imshow(normalize_minmax(mask), cmap='gray')
    plt.title(f"Máscara\nk={k}")
    plt.axis('off')

    plt.subplot(2, 5, i + 7)
    plt.imshow(sharpened, cmap='gray', vmin=0, vmax=255)
    plt.title(f"Highboost\nk={k}")
    plt.axis('off')

plt.tight_layout()
plt.show()
```

## Observações — Highboost Filtering

A aplicação do highboost filtering com diferentes valores do fator \( k \) mostra de forma clara como o ganho influencia a intensidade do realce na imagem final:

- **Máscara de detalhes (mask):**
  - **Comportamento visual:** A máscara representa os detalhes de alta frequência extraídos da imagem, isto é, principalmente bordas, contornos e pequenas variações de intensidade.
  - Como o kernel de borramento utilizado permanece o mesmo (**média 3x3**), a estrutura da máscara se mantém semelhante em todos os casos.
  - **Ruído:** A máscara também contém parte do ruído da imagem, já que ruídos são variações abruptas de intensidade e acabam sendo interpretados como detalhes.

- **Resultado final (highboost):**
  - **Comportamento visual:** O aumento do fator \( k \) torna o realce progressivamente mais forte.
    - Com **k = 1.0**, o resultado se aproxima do unsharp masking tradicional, com nitidez moderada e aparência mais natural.
    - Com **k = 1.5**, o reforço das bordas se torna mais evidente, melhorando a percepção de detalhes.
    - Com **k = 2.0**, a imagem apresenta realce forte, com contornos mais destacados.
    - Com **k = 3.0**, o efeito já se torna bastante agressivo, deixando a imagem com aparência artificial em algumas regiões.
  - **Bordas:** São progressivamente reforçadas à medida que o valor de \( k \) aumenta.
  - **Ruído:** Também é amplificado junto com os detalhes, especialmente para valores maiores de \( k \).
  - **Esmaecimento:** O método compensa o desfoque da imagem, restaurando e intensificando a nitidez.

- **Observação geral:**
  O highboost filtering permite controlar diretamente a intensidade do sharpening. Valores menores de \( k \) produzem resultados mais equilibrados, enquanto valores maiores aumentam bastante a nitidez, mas também podem exagerar ruídos e artefatos.

### Resumo do comportamento
- **k = 1.0:** realce moderado, mais natural.
- **k = 1.5:** nitidez mais evidente, bom equilíbrio visual.
- **k = 2.0:** realce forte, bordas bem destacadas.
- **k = 3.0:** efeito agressivo, com maior risco de amplificação de ruído e artificialidade.

Em geral, o highboost filtering é uma extensão do unsharp masking, com a vantagem de permitir ajuste fino da intensidade do realce por meio do fator \( k \).


## Detectores de Borda

### 1. Detector de bordas em cruz de Roberts

```python
#Função borda roberts
def roberts_edge(img):
    gx_kernel = np.array([
        [-1, 0],
        [0, 1]
    ], dtype=np.float32)

    gy_kernel = np.array([
        [0, -1],
        [1, 0]
    ], dtype=np.float32)

    gx = conv2d(img, gx_kernel, padding=True)
    gy = conv2d(img, gy_kernel, padding=True)

    magnitude = np.sqrt(gx**2 + gy**2)
    magnitude = np.clip(magnitude, 0, 255).astype(np.uint8)

    return gx, gy, magnitude
```

### 2. Detector de bordas de Sobel

```python
# Função borda sobel
def sobel_edge(img):
    gx_kernel = np.array([
        [-1, 0, 1],
        [-2, 0, 2],
        [-1, 0, 1]
    ], dtype=np.float32)

    gy_kernel = np.array([
        [-1, -2, -1],
        [ 0,  0,  0],
        [ 1,  2,  1]
    ], dtype=np.float32)

    gx = conv2d(img, gx_kernel, padding=True)
    gy = conv2d(img, gy_kernel, padding=True)

    magnitude = np.sqrt(gx**2 + gy**2)
    magnitude = np.clip(magnitude, 0, 255).astype(np.uint8)

    return gx, gy, magnitude
```

### 3. Detector de bordas laplacianas

```python
#Função borda laplacian
def laplacian_edge(img, kernel):
    lap = conv2d(img, kernel, padding=True)
    return lap, normalize_minmax(lap)
```

### Aplicando o filtro Sobel filtro de suavização

```python
# Função sobel com suavização
def sobel_after_smoothing(img, smooth_kernel):
    smoothed = conv2d(img, smooth_kernel, padding=True)
    gx, gy, magnitude = sobel_edge(smoothed)
    return smoothed, gx, gy, magnitude
```
